```python
# import packages needed
import math
import matplotlib.pyplot as plt
import numpy as np
import random
```

# Q1

```python
def simulator(time_list, thershold):
    random.seed(seed_value)
    phi = random.uniform(0, 1.)
    output_list = []
    for index, time_unit in enumerate(time_list):
        value = math.cos(2 * math.pi * ((time_unit / 12) + phi)) + (time_list[-1] - time_unit) * thershold
        output_list.append(np.around(value, decimals=4))
    return output_list
```

```python
seed_value = 100
time_list = [i for i in range(48)]
thershold = 0.06

output_list = simulator(time_list=time_list,
                       thershold=thershold)

print(output_list, '\n')

plt.plot(output_list)
plt.xlabel('time')
plt.ylabel('value')
plt.show()
```
