# Q1

(20%) Suppose the annual sales (in millions) of company A follow an AR(2) model: 
<br>
ùë¶ùë° = 5 + 1.1ùë¶ùë°‚àí1 ‚àí 0.5ùë¶ùë°‚àí2 + ùëéùë°, where ùúéùëé^2 = 2. 
<br><br>
(a) Show that the ùúì1 in the random shock form is also 1.1. 
<br>
(b) If the sales for 2005, 2006, and 2007 were 9, 11, and 10, respectively, forecast the sales for 2008 and 2009. 
<br>
(c) Calculate the 95% confidence interval of the 2008 forecast in (a). 
<br>
(d) If we now know the real sales of 2008 is 12, update your forecast for 2009.

# Answers
(a) the answer is in the exported cell below.
<br>
(b) according to our model, sales of 2008 and 2009 would be 11.0755, 12.7116. (result may varies each turn)
<br>
(c)
<br>
(d) according to our model, sales of 2009 would be 14.6644. (result may varies each turn)

```python
# (a)

# shared params
sample_size = 10000
arparams = np.array([1.1, -0.5])
maparams = np.array([0])

ar = np.r_[1, -arparams] # add zero-lag and negate
ma = np.r_[1, maparams] # add zero-lag

# error_terms = np.random.standard_t(df=6, size=sample_size)
# y_q2_arma = sm.tsa.arma_generate_sample(ar, ma, nsample=sample_size, distrvs=np.random.standard_t(), scale=1)

def np_random_standard_t():
    np.random.normal(0, np.sqrt(2))

y_q1_arma = sm.tsa.arma_generate_sample(ar, ma, nsample=sample_size, distrvs=np_random_standard_t())

y_q1_arma = [(y_q1_arma[i] + 5) for i in range(len(y_q1_arma))]

print(y_q1_arma[:10])


# fit the model with ma(1)
model_ma = ARMA(y_q1_arma , order=(0,1))
model_ma_fit = model_ma.fit()
print("\n\nMA(1) Model:\n", model_ma_fit.params)
print(model_ma_fit.summary())
```

```python
# (b) (c) (d)
arparams = np.array([1.1, 0.5])

def AR_model_predict(input_list:list, predict_length):
    output_list = []
    for i in range(predict_length):
        if i < (len(input_list)):
            output_list.append(input_list[i])
        else:
            value = 5 + 1.1 * output_list[i-1] - 0.5 * output_list[i-2] + np.random.normal(0, np.sqrt(2))
            output_list.append(np.around(value, decimals=4))
            
    return output_list
    
data_annual_sales = pd.DataFrame({
    "year":[2005, 2006, 2007],
    "sales":[9, 11, 10]
})

prediction = AR_model_predict(data_annual_sales.loc[:, "sales"].tolist(), 5)

print("Prediction of Sales (From 2005)", prediction)

data_annual_sales = pd.DataFrame({
    "year":[2005, 2006, 2007, 2008],
    "sales":[9, 11, 10, 12]
})

prediction = AR_model_predict(data_annual_sales.loc[:, "sales"].tolist(), 5)

print("Prediction of Sales Revised (From 2005)", prediction)
```

```python
# read in data
data_robot = pd.read_csv("./TSA HW06.robot.csv")
data_robot['date'] = list(data_robot.index)

data_robot_train = data_robot.iloc[:-5, :]
data_robot_test = data_robot.iloc[-5:, :]

model_ima = ARIMA(data_robot_train.iloc[:, 0], order=(0,1,1))
model_ima_fit = model_ima.fit()
print("\n\nIMA(1,1) Model:\n", model_ima_fit.params)
print(model_ima_fit.summary())
```

```python
# (a) (b)
# Some Reference: https://towardsdatascience.com/how-to-forecast-sales-with-python-using-sarima-model-ba600992fa7d
robot_ima_predictions = model_ima_fit.get_prediction(start=len(data_robot_train), end=(len(data_robot_train)+4))

pred_ci = robot_ima_predictions.conf_int(alpha=0.05)

ax = data_robot_test.iloc[:, 0].plot(label='observed')
robot_ima_predictions.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.95, figsize=(14, 4))
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)

print("Forecast Value:", robot_ima_predictions.predicted_mean.tolist())
print("95% Lower Bound:", pred_ci.iloc[:, 0].tolist())
print("95% Upper Bound:", pred_ci.iloc[:, 1].tolist())

plt.legend()
plt.show()
```

```python
# (c)

# fit the model with arma(1, 1)
model_ma = ARIMA(data_robot_train.iloc[:, 0], order=(1,0,1))
model_ma_fit = model_ma.fit()
print("\n\nIMA(1,1) Model:\n", model_ma_fit.params)
print(model_ma_fit.summary())

# (a) (b)
# Some Reference: https://towardsdatascience.com/how-to-forecast-sales-with-python-using-sarima-model-ba600992fa7d
robot_ma_predictions = model_ma_fit.get_prediction(start=len(data_robot_train), end=(len(data_robot_train)+4))

pred_ci = robot_ma_predictions.conf_int(alpha=0.05)

ax = data_robot_test.iloc[:, 0].plot(label='observed')
robot_ma_predictions.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.95, figsize=(14, 4))
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)

plt.legend()
plt.show()

print("Forecast Value:", robot_ma_predictions.predicted_mean.tolist())
print("95% Lower Bound:", pred_ci.iloc[:, 0].tolist())
print("95% Upper Bound:", pred_ci.iloc[:, 1].tolist())
```

```python
# (a)
data_airpass = pd.read_csv("./TSA HW08.airpass.csv")
data_airpass.loc[:, "log_airpass"] = np.log(data_airpass.loc[:, "airpass"])
print("Columns:", data_airpass.columns)

plt.plot(data_airpass.loc[:, "airpass"])
plt.title("Airpass")
plt.show()

plt.plot(data_airpass.loc[:, "log_airpass"])
plt.title("Airpass (Log-transformed)")
plt.show()
```

```python
# (b)
data_airpass_diff = data_airpass.diff(periods=1)

plt.plot(data_airpass_diff.loc[:, "log_airpass"])
plt.show()
```

```python
# (c)
data_airpass_diffseason = data_airpass_diff.diff(periods=4)

plt.plot(data_airpass_diffseason.loc[:, "log_airpass"])
plt.show()
```

```python
# (d)
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
data_airpass_acf = TSA.stattools.acf(data_airpass_diffseason.loc[:, "log_airpass"], nlags=10, fft=False)
fig = sm.graphics.tsa.plot_acf(data_airpass_diffseason.loc[:, "log_airpass"], lags=10, ax=ax1)
```

```python
# (e)
# fit data with SARIMA model

model=sm.tsa.statespace.SARIMAX(endog=data_airpass.loc[:, "log_airpass"],order=(0,1,1),seasonal_order=(0,1,1,12),trend='c',enforce_invertibility=False)
results=model.fit()

print(results.summary())
```

```python
# Some Reference: https://towardsdatascience.com/how-to-forecast-sales-with-python-using-sarima-model-ba600992fa7d
data_airpass_predictions = results.get_prediction(start=len(data_airpass), end=(len(data_airpass)+2))

pred_ci = data_airpass_predictions.conf_int(alpha=0.05)

ax = data_airpass.iloc[int(np.around((len(data_airpass)/5)*4, decimals=0)):, 1].plot(label='observed')
data_airpass_predictions.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.95, figsize=(14, 4))
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)

plt.legend()
plt.show()

print("Forecast Value:",data_airpass_predictions.predicted_mean.tolist())
print("95% Lower Bound:", pred_ci.iloc[:, 0].tolist())
print("95% Upper Bound:", pred_ci.iloc[:, 1].tolist())
```
