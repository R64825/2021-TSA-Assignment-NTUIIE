### Import packages

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import matplotlib.dates as dates
import statsmodels.api as sm
import itertools
from pmdarima.arima import auto_arima
from math import sqrt
from sklearn.metrics import mean_squared_error
```

### Function definition

```python
def Plot_Time_Series(poi, X, y, s , c):
    fig = plt.figure()
    plt.plot(X, y)
    plt.title(s+" "+ poi +" Changes at Wanhua, 2021")
    plt.xlabel('Time')
    plt.ylabel(poi)
    fig.set_figheight(6)
    fig.set_figwidth(10)

    plt.plot(X,y, color=c)
    plt.show()
```

```python
def OLS_Res(y):
    X = []
    for i in range(0, len(y)):
        X.append(i)
    reg = sm.OLS(y, X)
    result = reg.fit()
    # print result
    print("parameters: \n", result.params)
    print("\n", result.summary())
```

```python
def Auto_ARIMA(df):
    X = []
    for i in range(0, len(y)):
        X.append(i)
    # split train and test
    train = df[:int(train_Ratio*(len(df)))]
    test = df[int(train_Ratio*(len(df))):]
    
    Auto_ARIMA = auto_arima(X, trace=True, error_action='ignore', suppress_warnings=True).fit(train[poi])

    forecast = Auto_ARIMA.predict(n_periods=len(test))
    forecast = pd.DataFrame(forecast,index = test.index,columns=['Prediction'])

    #plot the predictions for validation set
    fig = plt.figure(figsize = (10,6))
    ax = fig.add_subplot()
    ax.plot(train.AQI, color = "blue", label="Train")
    ax.plot(test.AQI, color = "gray", label="test")
    ax.plot(forecast, color = "green", label="Prediction")
    ax.legend()
    plt.show()
    
    #calculate rmse
    rms = sqrt(mean_squared_error(test.AQI,forecast))
    print("rms:" + rms)
```

### settings

```python
input_path = r"C:\Users\TerryYang\(2021)_Preprocessed.csv"
train_Ratio = 0.8
poi = "AQI"
```

```python
# read dataset
df = pd.read_csv(input_path)
df['Date'] =pd.to_datetime(df.Date)
df = df.drop(columns = ['Unnamed: 0'])

df_Hour = df
df_Hour.head()
```

```python
df = df[["AQI", "Date"]]
df_Day = df.groupby([df['Date'].dt.date]).mean()
df_Day['Date'] = df_Day.index

df_Day.head()
```

```python
# visialize
X = df_Hour['Date']
y = df_Hour[poi]
Plot_Time_Series(poi, X, y, "Hourly" , "blue")

X = df_Day['Date']
y = df_Day[poi]
Plot_Time_Series(poi, X, y, "Daily", "green")
```

```python
y = df_Hour[poi]
OLS_Res(y)

y = df_Day[poi]
OLS_Res(y)
```
